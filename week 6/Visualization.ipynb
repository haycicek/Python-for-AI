```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HousingData.csv")
```

```python
display(df.head())
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
print(df.isnull().sum())
```

```python
df.to_csv("BostonHousing_Cleaned.csv", index=False)
```

```python
# Visualization - Relationship between RM and MEDV
plt.figure(figsize=(10, 6))
plt.scatter(df['RM'], df['MEDV'], alpha=0.5, color='green')
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median Price (MEDV)')
plt.title('Relationship Between Average Number of Rooms and Median Price')
plt.grid(True)
plt.show()
```

```python
# Visualization - Relationship between CRIM and MEDV
plt.figure(figsize=(10, 6))
plt.scatter(df['CRIM'], df['MEDV'], alpha=0.5, color='blue')
plt.xlabel('Crime Rate (CRIM)')
plt.ylabel('Median Price (MEDV)')
plt.title('Relationship Between Crime Rate and Median Price')
plt.grid(True)
plt.show()
```

```python
# Visualization - Relationship between TAX and MEDV
plt.figure(figsize=(10, 6))
plt.scatter(df['TAX'], df['MEDV'], alpha=0.5, color='red')
plt.xlabel('Property Tax Rate (TAX)')
plt.ylabel('Median Price (MEDV)')
plt.title('Relationship Between Property Tax Rate and Median Price')
plt.grid(True)
plt.show()
```
